```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentence=['It was many and many a year ago,In a kingdom by the sea,That a maiden there lived whom you may know,By the name of ANNABEL LEE']
```

```python
sentence
```

```python
voc_size = 10000
```

```python
onehot_repr=[one_hot(words,voc_size)for words in sentence] 
print(onehot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sentence_length=15
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sentence_length)
print(embedded_docs)
```

```python
dim = 35
```

```python
model=Sequential()
model.add(Embedding(voc_size,15,input_length=sentence_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs)[0])
```
